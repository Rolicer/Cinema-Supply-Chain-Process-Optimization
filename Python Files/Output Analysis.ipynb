```python
#general imports
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

#scipy for testing
import scipy
import scipy.stats
from scipy import stats
from scipy.stats import ttest_ind
import statsmodels.stats.api as sms

# sklearn for distribution transforming
from sklearn.preprocessing import QuantileTransformer

#for visualization
import matplotlib.pyplot as plt
import seaborn as sns

# for fitting the distribution
from fitter import Fitter, get_common_distributions, get_distributions

# other libraries
from pylab import rcParams
```

```python
path_to_data = 'C:/Users/dansi/OneDrive/Документы/Files/Data science/Bachelorseminar Cinedom/Data/Output analysis.csv'
df = pd.read_csv(path_to_data)
```

```python
data1 = df['OldtimeInSystem']
data2 = df['NewtimeInSystem']
stat, p = ttest_ind(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('There is significant evidance that the means are equal')
else:
	print('There is no significant evidance that the means are equal')

a = df['OldtimeInSystem']
print( '\n 95% Confidence Interval Status Quo model')
print(sms.DescrStatsW(a).tconfint_mean())

b = df['NewtimeInSystem']
print( '\n 95% Confidence Interval improved model')
print(sms.DescrStatsW(b).tconfint_mean())
```

```python
data3 = df['OldTotalWaiting']
data4 = df['NewTotalWaiting']
stat, p = ttest_ind(data3, data4)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('There is significant evidance that the means are equal')
else:
	print('There is no significant evidance that the means are equal')

a = df['OldTotalWaiting']
print( '\n 95% Confidence Interval Status Quo model')
print(sms.DescrStatsW(a).tconfint_mean())

b = df['NewTotalWaiting']
print( '\n 95% Confidence Interval improved model')
print(sms.DescrStatsW(b).tconfint_mean())
```
